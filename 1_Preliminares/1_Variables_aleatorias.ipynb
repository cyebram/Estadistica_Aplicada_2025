En este documento exploraremos las principales distribuciones de probabilidad continuas que son fundamentales para la inferencia estadística. Cada distribución será presentada con:
    
- Función de densidad de probabilidad (PDF)
- Función de distribución acumulativa (CDF)
- Esperanza matemática (media)
- Varianza
- Gráficas ilustrativas 

```python
# Importar las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import norm, uniform, expon, gamma, beta, chi2, t, f
import seaborn as sns
from matplotlib.patches import Rectangle
import warnings
warnings.filterwarnings('ignore')
    
# Configuración de gráficas
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 11
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.alpha'] = 0.3
    
print(Librerías importadas exitosamente)
```

```python
 # Distribución Normal
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Distribución Normal', fontsize=16, fontweight='bold')

# Diferentes medias, misma varianza
x = np.linspace(-8, 8, 1000)
medias = [0, 2, -1]
sigma = 1
for mu in medias:
    y = norm.pdf(x, mu, sigma)
    ax1.plot(x, y, label=f'μ={mu}, σ²={sigma**2}', linewidth=2)

ax1.set_title('Efecto de la Media (μ)')
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')
ax1.legend()
ax1.grid(True, alpha=0.3)
# Misma media, diferentes varianzas
mu = 0
sigmas = [0.5, 1, 2]
    
for sigma in sigmas:
    y = norm.pdf(x, mu, sigma)
    ax2.plot(x, y, label=f'μ={mu}, σ²={sigma**2}', linewidth=2)
    
ax2.set_title('Efecto de la Varianza (σ²)')
ax2.set_xlabel('x')
ax2.set_ylabel('f(x)')
ax2.legend()
ax2.grid(True, alpha=0.3)
    
# Función de distribución acumulativa
x_cdf = np.linspace(-4, 4, 1000)
y_cdf = norm.cdf(x_cdf, 0, 1)
ax3.plot(x_cdf, y_cdf, 'b-', linewidth=2, label='CDF')
ax3.axhline(y=0.5, color='r', linestyle='--', alpha=0.7, label='P(X ≤ 0) = 0.5')
ax3.axvline(x=0, color='r', linestyle='--', alpha=0.7)
ax3.set_title('Función de Distribución Acumulativa')
ax3.set_xlabel('x')
ax3.set_ylabel('F(x) = P(X ≤ x)')
ax3.legend()
ax3.grid(True, alpha=0.3)
    
# Área bajo la curva (probabilidades)
x_area = np.linspace(-4, 4, 1000)
y_area = norm.pdf(x_area, 0, 1)
ax4.plot(x_area, y_area, 'b-', linewidth=2, label='N(0,1)')
    
# Colorear área entre -1 y 1
x_fill = np.linspace(-1, 1, 100)
y_fill = norm.pdf(x_fill, 0, 1)
ax4.fill_between(x_fill, y_fill, alpha=0.3, color='red', label='P(-1 ≤ X ≤ 1) ≈ 0.68')
    
ax4.set_title('Regla 68-95-99.7 (Regla Empírica)')
ax4.set_xlabel('x')
ax4.set_ylabel('f(x)')
ax4.legend()
ax4.grid(True, alpha=0.3)
    
plt.tight_layout()
plt.show()
    
# Calcular probabilidades específicas
print("Probabilidades para N(0,1):")
#print("P(-1 ≤ X ≤ 1) = {norm.cdf(1, 0, 1) - norm.cdf(-1, 0, 1):.4f}\)
#print(f"P(-2 ≤ X ≤ 2) = {norm.cdf(2, 0, 1) - norm.cdf(-2, 0, 1):.4f}\)
#print(f"P(-3 ≤ X ≤ 3) = {norm.cdf(3, 0, 1) - norm.cdf(-3, 0, 1):.4f}\)
```
